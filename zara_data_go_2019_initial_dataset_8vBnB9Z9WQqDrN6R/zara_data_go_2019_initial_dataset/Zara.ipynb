```python
blocks["number_products"] = blocks["block_id"].apply(lambda row: df7.loc[df7["block_id"] == row]["product_id"].value_counts().count())
blocks["sales"] = blocks["block_id"].apply(lambda row: df7.loc[df7["block_id"] == row]["sales"].sum())
blocks["stock"] = blocks["block_id"].apply(lambda row: df7.loc[df7["block_id"] == row]["stock"].sum())
```

```python
blocks["values"] = blocks["sales"]
```

```python
for e in blocks["block_id"]:
    value = 0
    g = df7.loc[df7["block_id"] == e]
    for x in range(g.shape[0]):
        value += g["price"].iloc[x] * g["sales"].iloc[x]
    blocks["values"].loc[blocks["block_id"] == e] = value
```

```python
blocks["values"] = blocks["values"]/blocks["number_products"]
```
